### Research question
---
<p style="text-align: justify;">What is the influence of country-specific characteristics on the choice of supervision? This paper investigates the relation between each of the variables inflation, GDP per capita, bank capital to assets ratio and the Worldwide Governance Indicators on the choice of a supervisory institution.</p>

### Motivation
---

<p style="text-align: justify;">Financial supervision is a broad topic with several dimensions, such as who should act as a lender of last resort, or whether supervision should be on a national or supranational level. In our paper we choose to focus on the choice of the institution of which the supervisory role should be assigned to. This can be the central bank, a governmental agency or a combination of both. The choice of an institution and its power can be an important determinant for influences on factors such as bank profit efficiency. For example, [Gaganis and Pasiouras (2013)](http://www.sciencedirect.com/science/article/pii/S0378426613002094) find that bank profit efficiency decreases when a central bank supervises a higher number of financial sectors. However, the optimal choice of an institution ultimately depends on a diverse range of characteristics of the country that has to make this decision [(Eichengreen & Dincer, 2011)](http://www.nber.org/papers/w17401). Therefore, it is not obvious which institution should conduct supervision in a country, and most likely depends on macro-economic as well as social factors [(Ferran, 2011)](https://papers.ssrn.com/sol3/papers.cfm?abstract_id=1690523).</p>

<p style="text-align: justify;">The assignment of supervisory powers to a central bank or an agency could however have other optima for countries with different characteristics. Country A does not necessarily have to assign the supervision of the financial system to the same party as country B does, if the characteristics of country A and B differ substantially. This means that not either the option of a central bank, a governmental agency or both as supervisor necessarily has to be the right choice for all countries, it depends on the various characteristics of these countries.</p>

<p style="text-align: justify;">In order to find which type of countries would prefer a central bank or a governmental agency as their supervisor, we turn to the paper of [Eichengreen and Dincer (2011)](http://www.nber.org/papers/w17401). They have established a regression analysis in which they research which characteristics belong to countries choosing a central bank as their supervisor and in the same way which characteristics belong to countries with supervisory agencies. They have established these relationships by coding the structure of bank supervision as taking on different values when being assigned to an independent authority, a dependent governmental authority, and the central bank. They have used certain variables, as a representative for these characteristics, of each country that could be of influence to the choice of supervisor and regressed the data belonging to these variables on the values that have been assigned by the coding.</p>

<p style="text-align: justify;">Furthermore, we want to investigate whether countries with a higher bank capital to assets ratio are more likely to assign supervisory authority to central banks or governmental agencies. The reasoning behind this is that countries with a higher bank capital to assets ratio would in general have a more profound financial system, so it is interesting to research the influence of these capital ratios on the choice of supervisor.</p>

### Literature review
---
<p style="text-align: justify;">As we will focus on the same question that [Eichengreen and Dincer (2011)](http://www.nber.org/papers/w17401) have researched, we will only consider their paper in this literature review. We have summarized the findings of their findings in Table 1, in order to make clear what the impact of the variables is on the choice of supervisor. For the purpose of this paper, we only consider the relation of these variables to the choice between the central bank and a governmental agency, as where [Eichengreen and Dincer (2011)](http://www.nber.org/papers/w17401) more closely considered the independence of these agencies as well. In other words, if the coefficient of a certain variable on choosing as well an independent agency or a dependent agency is negative, the effect on this variable on agencies in general is negative. Hence, there would be a positive relation between this variable and choosing the central bank as supervisor, assuming the only choices for supervisors are central banks, independent agencies and dependent agencies. If one of the coefficients of a variable on choosing the independent or dependent agency would be negative and the other would be positive, the total effect of this variable on the choice of a governmental agency as a supervisor would be ambiguous. In that case, the effect of this variable on choosing a central bank as supervisor would be ambiguous as well.</p>

<p style="text-align: justify;">With these results, [Eichengreen and Dincer (2011)](http://www.nber.org/papers/w17401) show that there is a negative relation between GDP per capita and the choice of a central bank as a financial supervisor. In other words, less developed countries with lower GDP per capita would more likely assign the supervisory authority to their central banks. Another interesting result is that countries with higher political stability would be more prone to choosing central banks as their supervisor, even though countries with lower GDP per capita would choose a central bank as their supervisor as well. This is in contrary to the general consensus that within less developed countries there is a lower political stability, for instance due to a higher amount of corruption and politically-motivated violence and crime. Furthermore, they find a negative effect of regulatory quality on the choice of central bank. This in line with the theory that if the governmental has more regulatory quality, they will more likely trust the government and assign a governmental agency as the financial supervisor.</p>


### Method
---
<p style="text-align: justify;">In order to measure the relations of the different characteristics of countries on the choice of supervisor, in the same way as [Eichengreen & Dincer (2011)](http://www.nber.org/papers/w17401) conducted their research, it is necessary to gather data on the type of supervisor which is chosen and data on the variables tested for a significant amount of countries. Firstly, we established a list of countries of which the type of current supervisor was known. [Barth et al. (2013)](http://www.emeraldinsight.com/doi/abs/10.1108/17576381311329661) showed for 136 countries whether these countries have a central bank, governmental agency or a combination of both as supervisor. For these 136 countries, we took the data on the same variables as Eichengreen & Dincer. These are the six Worldwide Government Indicators, GDP per capita and Inflation rate. We choose to leave out the variables ‘German law’ and ‘UK law’, because there was no data on these variables available. Next to this, we had to omit 23 countries from the dataset, due to lacking data. This means that we are left with 113 countries in our dataset. Moreover, we added the variable ‘banking capital to assets ratio’. For many countries outside Europe, the bank capital to assets ratio for missing, so we had to omit these countries as well. Due to the fact that our data is now very much skewed towards Europe, we decided to focus on Europe only. The reasoning behind this is that countries in different continents can also have differents ways of assigning the supervisory role. And if we use a dataset that is filled with mostly European countries, we cannot make conclusions about the behavior in the entire world, which makes it better to focus on Europe only. Finally, we also have to omit the European countries of which we have no data on the past inflation. The data was collected from the World Bank database and we use the same definitions of these variables that the [Worldbank (2017)](http://info.worldbank.org/governance/wgi/index.aspx#doc) uses.</p>

<p style="text-align: justify;">We choose to conduct our research via OLS regression. In order to do this, we took the following steps. Firstly, we made three dummy variables SUPERCBi, SUPERAGENCYi and SUPERBOTHi which shows the type of supervision that is currently present in a country i. If a country i assigned the supervisory authorities to the central bank, then SUPERCBi = 1, and otherwise SUPERCBi = 0. The same reasoning goes for SUPERAGENCYi and SUPERBOTHi. Then, we use data on 9 variables to establish the influence of these variables on the dummy variable, thus the influence on whether they choose this type of supervision, for each country i. </p>

<p style="text-align: justify;">In order to conduct the OLS regression, we use the following regression formula with regard to choosing a central bank as financial supervisor:</p>

$$SUPERCB_{i} = \alpha +\beta _{1}ACCOUNT_{i} + \beta _{2}STABLE_{i} + \beta _{3}GOV_{i} + \beta _{4}QUALITY_{i} + \beta _{5}LAW_{i} + \beta _{6}CORRUPT_{i} + \beta _{7}CAP_{i} + \beta _{8}INFL_{i} + \beta _{9}GDP_{i} + \varepsilon$$

<p style="text-align: justify;">Using the same methodology for the regression of the variables on the choice of a governmental agency or both parties as financial supervisor gives the following equations:</p>

$$SUPERAGENCY_{i} = \alpha +\beta _{1}ACCOUNT_{i} + \beta _{2}STABLE_{i} + \beta _{3}GOV_{i} + \beta _{4}QUALITY_{i} + \beta _{5}LAW_{i} + \beta _{6}CORRUPT_{i} + \beta _{7}CAP_{i} + \beta _{8}INFL_{i} + \beta _{9}GDP_{i} + \varepsilon$$

$$SUPERBOTH_{i} = \alpha +\beta _{1}ACCOUNT_{i} + \beta _{2}STABLE_{i} + \beta _{3}GOV_{i} + \beta _{4}QUALITY_{i} + \beta _{5}LAW_{i} + \beta _{6}CORRUPT_{i} + \beta _{7}CAP_{i} + \beta _{8}INFL_{i} + \beta _{9}GDP_{i} + \varepsilon$$

<p style="text-align: justify;">Next to this, we conducted a sensitivity analysis to test whether the addition of other variables influences the effect of one variable to the choice of the central bank or agency as the supervisor. Since GDP and the Worlbank Government Indicators are highly correlated, we expect that excluding these indicators in the equation does not alter the effect of GDP on the choice of supervisor much. Therefore, we use the following regression equations to conduct the sensitivity analysis:</p>

$$SUPERCB_{i} = \alpha + \beta _{1}GDP_{i} + \varepsilon$$

$$SUPERAGENCY_{i} = \alpha + \beta _{1}GDP_{i} + \varepsilon$$

$$SUPERBOTH_{i} = \alpha + \beta _{1}GDP_{i} + \varepsilon$$

<p style="text-align: justify;">We find that less developed countries with lower GDP per capita ratios more likely opt for a central bank as supervisor. ### Results
---

```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
import statsmodels.api as sm
import pandas as pd
from pandas.tools.plotting import scatter_matrix
```

```python
#loading the dataset into the notebook
xl = pd.ExcelFile("Data Final Assignment.xlsx")
unfiltered_data = xl.parse("Sheet1")

#filtering the countries that do not have data with regard to inflation
data = unfiltered_data[~np.isnan(unfiltered_data["infl"])]

#adapting the column names to make their content more concrete
data.columns = "Country","Accountability","Political stability","Government effectiveness","Regulatory quality","Rule of law","Control of corruption","GDP per capita","Past inflation","Bank capital to assets ratio","SUPERCB","SUPERAGENCY","SUPERBOTH"
data
```

```python
#creating the OLS regression for SUPERCB, including all variables
y = data['SUPERCB']
x = data[['Accountability','Political stability','Government effectiveness','Regulatory quality','Rule of law','Control of corruption','GDP per capita','Past inflation','Bank capital to assets ratio']]
x = sm.add_constant(x, prepend=True)
results = sm.OLS(endog=y,exog=x).fit()

print results.summary()
```

```python
#creating the OLS regression for SUPERAGENCY, including all variables
y = data['SUPERAGENCY']
x = data[['Accountability','Political stability','Government effectiveness','Regulatory quality','Rule of law','Control of corruption','GDP per capita','Past inflation','Bank capital to assets ratio']]
x = sm.add_constant(x, prepend=True)
results = sm.OLS(endog=y,exog=x).fit()

print results.summary()
```

```python
#creating the OLS regression for SUPERBOTH, including all variables
y = data['SUPERBOTH']
x = data[['Accountability','Political stability','Government effectiveness','Regulatory quality','Rule of law','Control of corruption','GDP per capita','Past inflation','Bank capital to assets ratio']]
x = sm.add_constant(x, prepend=True)
results = sm.OLS(endog=y,exog=x).fit()

print results.summary()
```

<p style="text-align: justify;">In Table 2 we show the output of the three separate regressions, which is the effect of each variable on choosing a central bank, governmental agency or a combination of both as a supervisor. The conclusion that can immediately be drawn from this table is that except for one variable, namely bank capital to assets ratio, the effects on choosing a central bank or a governmental agency move in opposite directions. This is in line with our expectations, because choosing a central bank or governmental agency are contrary choices. Furthermore, we see that the effects on choosing both parties for merely all variables are in the same direction as the effect on choosing only a governmental agency, and thereby in contrast to the effect on the central bank, except for the variables past inflation, political stability and regulatory quality. This suggests that there is more diversion in characteristics between countries that choose central banks as their lone supervisor and those that choose both than between countries that choose a governmental agency as their lone supervisor and those that choose both. Hence, countries that choose either the governmental agency or both parties as supervisor are more similar.</p>

<p style="text-align: justify;">We find that five variables have a positive effect on the choice of a central bank as the financial supervisor, namely GDP per capita, past inflation, political stability, accountability and government effectiveness. For agency supervision we also find positive relations for the outcomes of five variables, namely GDP per capita, rule of law, control of corruption, regulatory quality and bank capital to assets ratio. For both, we find positive effects for past inflation, rule of law, control of corruption, political stability and bank capital to assets ratio.</p>

<p style="text-align: justify;">However, we find that most of our results are highly insignificant. Therefore we cannot make certain conclusions based on these results. One possible cause for this is that our variables, especially the Worldwide Government Indicators, are highly correlated. We can test this by doing a correlation analysis. </p>

```python
#checking for a high correlation between the used variables
np.corrcoef(data[['Accountability','Political stability','Government effectiveness','Regulatory quality','Rule of law','Control of corruption','GDP per capita','Past inflation','Bank capital to assets ratio']])
```

<p style="text-align: justify;">We indeed find that the correlation between our variables is very high, which leads to insignificant results when we regress them together in the same OLS regression. To make sure that this could be the reason, we conduct a sensitivity analysis for each of the three dependent variables, taking GDP per capita as the only independent variable. </p> 

```python
#conducting the sensitivity analysis with the variable 'GDP per capita' for SUPERCB
y = data['SUPERCB']
x = data[['GDP per capita']]
x = sm.add_constant(x, prepend=True)
results = sm.OLS(endog=y,exog=x).fit()

print results.summary()
```

```python
#conducting the sensitivity analysis with the variable 'GDP per capita' for SUPERAGENCY
y = data['SUPERAGENCY']
x = data[['GDP per capita']]
x = sm.add_constant(x, prepend=True)
results = sm.OLS(endog=y,exog=x).fit()

print results.summary()
```

```python
#conducting the sensitivity analysis with the variable 'GDP per capita' for SUPERBOTH
y = data['SUPERBOTH']
x = data[['GDP per capita']]
x = sm.add_constant(x, prepend=True)
results = sm.OLS(endog=y,exog=x).fit()

print results.summary()
```

<p style="text-align: justify;">The regressions indeed show significant results for the effect of GDP per capita on the choice of a central bank as supervisor and for choosing both a Central Bank and governmental agency as supervisors. In our initial regression for the central bank though, GDP per capita showed an insignificant effect. So excluding the control variables moved the effect of GDP per capita on the choice of the central bank as the supervisor from statistically insignificant to statistically significant results. However, we now suspect what the reason is why our research showed insignificant results but we cannot just leave the highly correlated control variables out of the regression. We cannot simply regress one variable at the time for each dependent variable, just in order to get significant results, because these significant results would not give a true indication of the effect without the control variables.</p>

### Discussion

Inflation, accountability and government effectiveness gave ambiguous results in the paper of [Eichengreen and Dincer (2011)](http://www.nber.org/papers/w17401) and could therefore not give the same results as in our study, where we only have positive or negative effects. </p>

<p style="text-align: justify;">Our regression showed positive results on choosing a central bank as supervisor from the variables GDP per capita, past inflation, political stability, accountability and government effectiveness. This could mean that countries that are economically growing inflation-wise and more politically stable are more likely to opt for a central bank as their supervisor, because they have more trust in their central bank dealing with the extra power. The choice of a governmental agency is strengthened by having a higher score on GDP per capita, rule of law, control of corruption, regulatory quality and bank capital to assets ratio. This could mean that countries that need to have more control in law, corruption and banking capital will more likely choose a governmental agency as the financial supervisor. One possible explanation for this could be that they feel that the central bank will misuse the extra power it has. </p>

<p style="text-align: justify;">The low coefficients of some variables as discussed can be explained by the correlation coefficients between the World Government Indicators as discussed before. Because of these high correlations, it was not expected that these factors have diverse effects on the choice of supervision. </p>
